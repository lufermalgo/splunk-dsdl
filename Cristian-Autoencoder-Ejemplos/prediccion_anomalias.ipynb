```python
deteccion=dfk1['anomalia'].shift(5).fillna(0)
```

```python
deteccion=deteccion[:508320].copy()
```

```python
def images_gen(X):
    img_=[]
    for i in range(X.shape[0]):
        img_.append(X[i].T/X[i].T.max(axis=0))
    return np.array(img_)
```

```python
Xtemp=images_gen(turnos_)
```

```python
Xtemp.shape
```

```python
Xtemp=Xtemp.reshape((508320,8))
```

```python
dfk1['anomalia'].value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_desbalanceado_rf=RandomForestClassifier().fit(Xtemp,deteccion)
```

```python
from sklearn.metrics import f1_score
```

```python
resultado=modelo_desbalanceado_rf.predict(Xtemp)
```
