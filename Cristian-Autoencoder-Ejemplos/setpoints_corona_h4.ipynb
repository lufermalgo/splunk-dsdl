### Heuristica

```python
def verificacion_valor(sigma,limite):
    return 0
```

```python
def heuristica_temperatura(setpoint,reduccion,limite_maximo_sp,limite_maximo_in, temperatura,delta_esperado):
    nuevo_setpoint=setpoint-reduccion
    valor_esperado=nuevo_setpoint+delta_esperado
    ls_sp=valor_esperado+limite_maximo_sp
    li_sp=valor_esperado-limite_maximo_in
    sigma_sp=temperatura-ls_sp
    sigma_li=li_sp-temperatura
    if sigma_sp>0:
        return ls_sp-sigma_sp
    if sigma_li>0:
        return li_sp+sigma_li
    else:
        return temperatura
```

```python
def comportamiento_esperado(X,setpoins_,reducciones,lsup,linf,queantil_termo):
    t_=[]
    for l in range(26):
        T_=[]
        for i in X[:,l]:
            a=heuristica_temperatura(setpoins_[l],reducciones[l],lsup[l],linf[l],i,queantil_termo[l])
            T_.append(heuristica_temperatura(setpoins_[l],reducciones[l],lsup[l],linf[l],a,queantil_termo[l]))
        t_.append(T_)
    return np.array(t_).T
```
